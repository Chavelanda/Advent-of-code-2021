```python
import numpy as np
```

```python
crabs = np.loadtxt('input.txt', delimiter=',', dtype=np.int_)
crabs
```

```python
max_pos = np.max(crabs)
max_pos
```

```python
min_pos = np.min(crabs)
min_pos
```

```python
fuels = np.zeros(max_pos, dtype=np.int_)

for i in range(max_pos):
    c_crabs = np.copy(crabs)
    c_crabs -= i
    c_crabs = np.abs(c_crabs)
    fuels[i] = np.sum(c_crabs)
```

```python
np.min(fuels)
```

```python
def calc_fuel(dist):
    return dist*(dist+1)/2
```

```python
vect_calc_fuel = np.vectorize(calc_fuel)

fuels = np.zeros(max_pos, dtype=np.int_)

for i in range(max_pos):
    c_crabs = np.copy(crabs)
    c_crabs -= i
    c_crabs = np.abs(c_crabs)
    c_crabs = vect_calc_fuel(c_crabs)
    fuels[i] = np.sum(c_crabs)
```

```python
np.min(fuels)
```
