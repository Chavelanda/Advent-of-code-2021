```python
import pandas as pd
import numpy as np
```

```python
filename = open('input.txt', "r")
heights = []
gr = filename.readline().strip()
while gr:
    gr = list(map(int,str(gr)))
    heights.append(gr)
    gr = filename.readline().strip()

heights = np.array(heights, dtype=np.int_)
heights
```

```python
print(heights.shape)
ph = np.pad(heights, 1, constant_values=9)
ph.shape
```

```python
risk = 0
basin_points = ([], [])
basins = {}
low_points_n = 0
for i in range(1, 101):
    for j in range(1, 101):
        n = ph[i, j]
        if n < ph[i, j-1] and n < ph[i-1, j] and n < ph[i, j+1] and n < ph[i+1, j]:
            risk += n + 1
            basin_points[0].append(i)
            basin_points[1].append(j)
            basins[low_points_n] = [(i, j)]
            low_points_n += 1
print(risk)
print(basin_points)
print(basins)
```

```python
def near(i, j, basin):
    for ib, jb in basin:
        if (i == ib and (j == jb +1 or j == jb - 1)) or (j == jb and (i == ib +1 or i == ib - 1)):
            return True
    return False

def assign(i, j):
    for basin in basins:
        if (i, j) not in basins[basin] and near(i, j, basins[basin]):
            basins[basin].append((i, j))
            basin_points[0].append(i)
            basin_points[1].append(j)
            return 1
    return 0
```

```python
phc = np.copy(ph)
assigned = -1
while assigned != 0:
    assigned = 0
    phc[basin_points] = 9
    for i in range(1, 101):
        for j in range(1, 101):
            if phc[i, j] != 9:
                assigned += assign(i, j)
    print(assigned)
```

```python
lengths = np.zeros((len(basins),), dtype=np.int_)

for basin in basins:
    lengths[basin] = len(basins[basin])
```

```python
sorted_lengths = np.sort(lengths)
sorted_lengths
```
